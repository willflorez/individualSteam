```python
import pandas as pd
import json
%load_ext autoreload
%autoreload 2
import utils
import warnings
warnings.filterwarnings("ignore")
import ast
```

```python
path = 'australian_users_items.json'

rows2 = []

with open (path, encoding='utf-8') as f:
    for line in f.readlines():
        rows2.append(ast.literal_eval(line))

df_items = pd.DataFrame(rows2)

unpacked_data = []
for index, row in df_items.iterrows():
    user_id = row['user_id']
    items_count = row['items_count']
    steam_id = row['steam_id']
    user_url = row['user_url']
    for review_dict in row['items']:
        review_dict['user_id'] = user_id
        review_dict['items_count'] = items_count
        review_dict['steam_id'] = steam_id
        review_dict['user_url'] = user_url
        unpacked_data.append(review_dict)

df_items = pd.DataFrame(unpacked_data)
```
